**Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.**

```python
chipotle = pd.read_sql('select * from orders', env.get_db_url(env.username, env.hostname, env.password, 'chipotle'))
chipotle.head()

chipotle['item_price'] = chipotle['item_price'].replace('[$]', '', regex=True).astype(float)
chipotle['total_revenue'] = chipotle['item_price']

grouped = chipotle.groupby('item_name').agg({'quantity': 'sum', 'total_revenue': 'sum'})

top_items = grouped.sort_values('quantity', ascending=False).head(4)

top_items['total_revenue'].plot(kind='bar', color='skyblue')

plt.title('Top 4 Popular Items and Revenue Produced')
plt.xlabel('Item Name')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

**Load the sleepstudy dataset from pydataset and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.**

```python
df5 = data('sleepstudy') # load the dataset and store it in a variable
df5.head()
```

```python
df5.Subject = df5.Subject.astype(str)
plt.figure(figsize=(12,8))
sns.lineplot(data=df5, x='Days', y='Reaction', hue='Subject', alpha=0.6)
average = df5.groupby('Days')[['Reaction']].mean()
sns.lineplot(data=average,x=average.index, y='Reaction', color='Black')
```
