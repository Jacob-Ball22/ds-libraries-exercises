```python
import matplotlib.pyplot as plt
```

**Create and label 4 separate charts with the following equations (choose a range for x that makes sense):**

* y=√x

```python
from math import sqrt
x = list(range(0,51))
y = [sqrt(n) for n in x]
plt.scatter(x,y)
plt.title('$y = √x$')
plt.xlabel('$x$')
plt.ylabel('$√x$')
```

* y=x^3

```python
x = list(range(-50,51))
y = [n ** 3 for n in x]
plt.scatter(x,y)
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
```

* y=2x

```python
x = list(range(-50,51))
y = [2 ** n for n in x]
plt.scatter(x,y)
plt.title('$y = 2x$')
plt.xlabel('$x$')
plt.ylabel('$2x$')
```

* y=1/(x+1)

```python
x = list(range(0,51))
y = [1/(n + 1) for n in x]
plt.scatter(x,y)
plt.title('$y = 1/(x + 1)$')
plt.xlabel('$x$')
plt.ylabel('$1/(x + 1)$')
```

**Combine the figures you created in the last step into one large figure with 4 subplots.**

```python
x1 = list(range(0,51))
y1 = [sqrt(n) for n in x1]

x2 = list(range(-50,51))
y2 = [n ** 3 for n in x2]

x3 = list(range(-50,51))
y3 = [2 ** n for n in x3]

x4 = list(range(0,51))
y4 = [1/(n + 1) for n in x4]

plt.subplot(221)
plt.plot(x1,y1)
plt.title('$y = √x$')

plt.subplot(222)
plt.plot(x2,y2)
plt.title('$y = x^3$')

plt.subplot(223)
plt.plot(x3,y3)
plt.title('$y = 2x$')

plt.subplot(224)
plt.plot(x4,y4)
plt.title('$y = 1/(x + 1)$')

plt.tight_layout(pad=1)
```

**Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.**

```python
x1 = list(range(0,51))
y1 = [sqrt(n) for n in x1]

x2 = list(range(-50,51))
y2 = [n ** 3 for n in x2]

x3 = list(range(-50,51))
y3 = [2 ** n for n in x3]

x4 = list(range(0,51))
y4 = [1/(n + 1) for n in x4]

plt.subplot(221)
plt.plot(x1,y1, color = 'red')
plt.title('$y = √x$')

plt.subplot(222)
plt.plot(x2,y2, color = 'blue')
plt.title('$y = x^3$')

plt.subplot(223)
plt.plot(x3,y3, color = 'green')
plt.title('$y = 2x$')

plt.subplot(224)
plt.plot(x4,y4, color = 'purple')
plt.title('$y = 1/(x + 1)$')

plt.tight_layout(pad=1)
```

**Use matplotlib to plot the following equation:**

* y=x^2−x+2

* You'll need to write the code that generates the x and y points.

* Add an anotation for the point 0, 0, as the origin.

```python
x = list(range(-20,21))
y = [(n**2- n + 2) for n in x]
plt.scatter(x,y)
plt.annotate('(0,0)', xy=(0, 0), xytext=(0, 50),
             arrowprops={'facecolor': 'red'})
```
